**Now you can see that there are certain columns which are not required to predict the breast cancer like the id col and unnamed:32 col   **

```python

df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
df.head(5)
```

**assigning the features columns to X and the target column to y**

```python
y=df['diagnosis']
X=df.drop('diagnosis',axis=1)
```

```python
y
```

```python
X.head(5)
```

**Replacing the values of Y from M and B to 1 and 0**

```python
y=y.replace(['M'],1)
y=y.replace(['B'],0)
y
```

```python
X.dtypes
```

****Now we will split the given data as training data and testing data using sklearn****

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
N, D = X_train.shape
```

**Now we will create a Keras 2.0 model for this classification problem **

```python

import tensorflow as tf 
from tensorflow import keras
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(D,)),
  tf.keras.layers.Dense(1, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
    ])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
r = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=25)
```

```python
model.history.history
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss
```

```python
model_loss.plot()
```

**Model Evaluation **

```python
predictions = model.predict(X_test).round()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
